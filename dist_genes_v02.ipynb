### Dataframe de Genes

##### Obtenemos los registros de tipo 'gene' y 'pseudogene'

```python
data_genes = df[(df['type']=='gene')|(df['type']=='pseudogene')]
```

```python
data_genes
```

##### Eliminamos del dataframe anterior los genes con ID único

```python
# Obtener el recuento de cada valor
value_counts = data_genes['seq_id'].value_counts()
```

```python
# Seleccionar los valores donde el recuento es igual a 1
to_remove = value_counts[value_counts == 1].index
```

```python
# Mantener las filas donde la columna de ID no esté en to_remove
data_genes_2 = data_genes[~data_genes.seq_id.isin(to_remove)]
```

```python
data_genes_2
```

```python
# No se observan valores únicos de IDs
data_genes_2['seq_id'].value_counts()
```

##### Ordenamos el dataframe primero por strand (+,-), luego por 'seq_id, y por último 'start' y 'end'

```python
data_genes_2.sort_values(['strand', 'seq_id','start', 'end'], inplace = True, ascending=True)
```

##### Reseteamos el índice del dataframe

```python
data_genes_2.reset_index(drop=True, inplace=True)
```

##### Eliminamos columnas innecesarias

```python
data_genes_2.drop(columns=['source', 'type', 'score', 'phase', 'attributes'], inplace = True)
```

##### Eliminamos la primera fila del dataframe 'data_genes_2' y luego reseteamos el índice

```python
data_genes_sin_fila_cero = data_genes_2.drop(0, axis=0)
data_genes_sin_fila_cero.reset_index(drop=True, inplace=True)
```

##### Ciclo o secuencia en el que vamos almacenando en una columna denominada 'dist_genes' la distancia entre genes
Recordemos que la distancia entre genes será igual al valor absoluto entre (data_genes_sin_fila_cero['start'] - data_genes_2['end']). Debemos tener en cuenta que los calculos de resta no se hagan entre celdas de scaffolds diferentes.

```python
cont = 0
for i in range(len(data_genes_2[['strand', 'seq_id']].drop_duplicates())):
    cont = 0
    strand_i = data_genes_2[['strand', 'seq_id']].drop_duplicates().iloc[i].strand
    seq_id_i = data_genes_2[['strand', 'seq_id']].drop_duplicates().iloc[i].seq_id
    for j,k,m,n,h,z in zip(data_genes_2['strand'], data_genes_2['seq_id'], data_genes_2['end'], data_genes_sin_fila_cero['start'], data_genes_sin_fila_cero['strand'], data_genes_sin_fila_cero['seq_id']):
        if strand_i == j and seq_id_i == k and strand_i == h and seq_id_i == z:
            data_genes_2.at[cont,'dist_genes']= abs(n-m)
        cont = cont + 1
```

##### Completamos con ceros los valores nulos

```python
data_genes_2['dist_genes'].fillna(0, inplace=True)
```

```python
data_genes_2
```

##### Agrupamos por 'seq_id' y obtenemos el promedio por cada uno de los ID

```python
data_genes_3 = data_genes_2[['seq_id', 'dist_genes']].groupby('seq_id').mean().reset_index()
```

```python
data_genes_3
```

```python
# Imprimimos la media de la columna
print(data_genes_3['dist_genes'].mean())
```

### Dataframe de Scaffolds únicos

##### Obtenemos los registros de tipo 'region'

```python
data_regiones = df[df['type']=='region']
```

```python
data_regiones
```

##### Se observan que los IDs tienen counts únicos

```python
data_regiones['seq_id'].value_counts()
```

```python
data_regiones.reset_index(drop=True, inplace=True)
```

```python
data_regiones
```

##### Incorporamos la lista de Scaffolds removidos del dataframe de genes en un dataframe llamado 'to_remove_serie'

```python
to_remove_serie = pd.DataFrame(to_remove.tolist())
```

```python
to_remove_serie
```

```python
to_remove_serie.shape
```

##### Unimos el dataframe anterior con 'data_genes', para poder ver los datos de inicio y fin del gen

```python
scaffolds_unicos_gene = to_remove_serie.merge(data_genes, left_on =0, right_on = 'seq_id', how = 'left').drop(columns = [0])
```

```python
scaffolds_unicos_gene
```

##### Se observa que, evidentemente, los valores siguen siendo únicos

```python
scaffolds_unicos_gene['seq_id'].value_counts()
```

##### Eliminamos columnas innecesarias para los dataframes de 'scaffolds_unicos_gene' y 'data_regiones'

```python
scaffolds_unicos_gene.drop(columns=['source', 'type', 'score', 'phase', 'attributes'], inplace = True)
```

```python
scaffolds_unicos_gene
```

```python
data_regiones.drop(columns=['source', 'type', 'score', 'phase', 'attributes'], inplace = True)
```

```python
data_regiones
```

##### Unimos los dataframes de scaffolds con los de 'data_regiones', con el objetivo de tener en un único dataframe el fin e inicio del gen y de las regiones

```python
scaffolds_unicos_gene_2 = scaffolds_unicos_gene.merge(data_regiones, left_on ='seq_id', right_on = 'seq_id', how = 'left').drop(columns=['strand_y'])
```

```python
scaffolds_unicos_gene_2
```

##### Renombramos las columnas

```python
scaffolds_unicos_gene_2.rename(columns={'start_x':'start_gene', 'end_x':'end_gene', 'strand_x':'strand', 'start_y':'start_region', 'end_y':'end_region'}, inplace = True)

scaffolds_unicos_gene_2.columns
```

```python
scaffolds_unicos_gene_2
```

##### Hallamos el inicio del gen 'd0', que será evidentemente igual a 'start_gene'

```python
scaffolds_unicos_gene_2['d0'] = scaffolds_unicos_gene_2['start_gene']
```

```python
scaffolds_unicos_gene_2
```

##### Hallamos la distancia 'd2', que será igual al fin de la region menos el fin del gen

```python
scaffolds_unicos_gene_2['d2'] = scaffolds_unicos_gene_2['end_region'] - scaffolds_unicos_gene_2['end_gene']
```

```python
scaffolds_unicos_gene_2
```

##### La distancia conocida mínima será igual a la suma entre 'd0' y 'd2'

```python
scaffolds_unicos_gene_2['dist_conocida'] = scaffolds_unicos_gene_2['d0'] + scaffolds_unicos_gene_2['d2']
```

```python
scaffolds_unicos_gene_2
```
